```python
##
##Tables based on Rojan UML

#new = old.filter(['A','B','D'], axis=1)

unique_runners = df3.drop_duplicates(subset=['name', 'birth_date'], keep='first')
unique_runners.reset_index()


df3['team'] = df3['team'].fillna('Individual')
df3 = df3[df3['age_class'].notna()]
df3
```

```python
#Creating the runners table with runner_index as key


runners_table = unique_runners.filter(['name', 'sex', 'nation', 'birth_date'])
#runners_table2 = runners_table.reset_index()
#runners_table = runners_table2.filter(['name', 'sex', 'nation', 'birth_date'])
runners_table.insert(0, "runner_index", range(1, len(runners_table) + 1), True)
runners_table = runners_table.reset_index()
runners_table2 = runners_table.drop(runners_table.columns[0],axis = 1)
#runners_table['runner_index'] = range(1, len(runners_table) + 1)

runners_table2
```

```python
#Event


#unique_events = df3.drop_duplicates(subset=['event','event_year','distance'], keep='first')
unique_events = df3.drop_duplicates(subset=['event'], keep='first')
unique_events

#unique_events = unique_events.filter(['event','event_year', 'distance'])
unique_events = unique_events.filter(['event'])
unique_events.insert(0, "event_index", range(1, len(unique_events) + 1), True)
unique_events = unique_events.reset_index()
unique_events2 = unique_events.drop(unique_events.columns[0],axis = 1)

unique_events2
```

```python
unique_races = df3.drop_duplicates(subset=['event','event_year','distance'], keep='first')

unique_races = unique_races.filter(['event','event_year', 'distance'])
unique_races.insert(0, "race_index", range(1, len(unique_races) + 1), True)

for i in range(len(unique_races)):
    if unique_races.iloc[i,1] == 'dia-do-pai':
        unique_races.iloc[i,1] = 1
    if unique_races.iloc[i,1] == 'family-race':
        unique_races.iloc[i,1] = 2
    if unique_races.iloc[i,1] == 'mar':
        unique_races.iloc[i,1] = 3
    if unique_races.iloc[i,1] == 'maratona':
        unique_races.iloc[i,1] = 4
    if unique_races.iloc[i,1] == 'meia_maratona':
        unique_races.iloc[i,1] = 5
    if unique_races.iloc[i,1] == 'sao-joao':
        unique_races.iloc[i,1] = 6
    if unique_races.iloc[i,1] == 'sao-silvestre':
        unique_races.iloc[i,1] = 7
        
unique_races = unique_races.reset_index()
unique_races2 = unique_races.drop(unique_races.columns[0],axis = 1)

unique_races2

```

```python
#Team

table_team = df3.drop_duplicates(subset=['team'], keep='first')


table_team = table_team.filter(['team'])
table_team.insert(0, "team_index", range(1, len(table_team) + 1), True)
table_team = table_team.reset_index()
table_team2 = table_team.drop(table_team.columns[0],axis = 1)

table_team2
```

```python
unique_age_class = df3.drop_duplicates(subset=['age_class'], keep='first')
unique_age_class = unique_age_class.filter(['age_class'])
unique_age_class.insert(0, "age_class_index", range(1, len(unique_age_class) + 1), True)

unique_age_class = unique_age_class.reset_index()
unique_age_class2 = unique_age_class.drop(unique_age_class.columns[0],axis = 1)

unique_age_class2
```

```python
reservation = df3.drop_duplicates(subset=['team', 'bib', 'name', 'event','age_class','birth_date','event_year'], keep='first')
reservation = reservation.filter(['team', 'bib', 'name', 'event','age_class','birth_date','event_year','place_in_class','net_time', 'official_time','place'])


#reservation.insert(0, "reservation_index", range(1, len(reservation) + 1), True)

reservation = reservation.reset_index()
reservation2 = reservation.drop(reservation.columns[0],axis = 1)

reservation2
```

```python
for i in range(len(reservation2)):
    team = reservation2.iloc[i,0]
    index_team = table_team2.loc[table_team2['team'] == team]['team_index']
    reservation2.iloc[i,0] = index_team
    
reservation2
```

```python
for i in range(len(reservation2)):
    event = reservation2.iloc[i,3]
    index_event = unique_events2.loc[unique_events2['event'] == event]['event_index']
    reservation2.iloc[i,3] = index_event
    
reservation2
    
```

```python
race_index = [None] * (len(reservation2.index))

for i in range(len(reservation2)):
    event = reservation2.iloc[i,3]
    event_year = reservation2.iloc[i,6]
    index_race = unique_races2.loc[(unique_races2['event'] == event) & (unique_races2['event_year'] == event_year)]['race_index']
    race_index[i] = index_race.iloc[0]
    
reservation2.insert(1, "race_index", race_index , True)
reservation2  
```

```python
runner_index = [None] * (len(reservation2.index))

for i in range(len(reservation2)):
    runner = reservation2.iloc[i,4]
    birth_date = reservation2.iloc[i,7]
    index_runner = runners_table2.loc[(runners_table2['birth_date'] == birth_date) & (runners_table2['name'] == runner)]['runner_index']
    runner_index[i] = index_runner.iloc[0]
    
reservation2.insert(1, "runner_index", runner_index , True)
reservation2
```

```python
reservation2 = reservation2[reservation2['age_class'].notna()]
age_class_index = [None] * (len(reservation2.index))

for i in range(len(reservation2)):
    age_class = reservation2.iloc[i,7]
    index_age_class = unique_age_class2.loc[unique_age_class2['age_class'] == age_class]['age_class_index']
    reservation2.iloc[i,7] = index_age_class.iloc[0]

reservation2
```
